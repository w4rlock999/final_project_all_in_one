```python
from langfuse import Langfuse

LANGFUSE_SECRET_KEY="sk-lf-6c7f1d9c-9dc6-4061-bb81-a77b9b3d5ffe"
LANGFUSE_PUBLIC_KEY="pk-lf-dd1577e9-be42-4bc0-9cb6-73fd709f3cf9"
LANGFUSE_HOST="http://localhost:3000"

langfuse = Langfuse(
    secret_key=LANGFUSE_SECRET_KEY,
    public_key=LANGFUSE_PUBLIC_KEY,
    host=LANGFUSE_HOST
)
```

```python
from datetime import datetime
import json

timestamp = datetime(2025, 5, 3, 18, 32, 24, 690000)

traces = langfuse.fetch_traces(from_timestamp=timestamp)

# Create a list to store all traces with their observations
traces_with_observations = []

# Process each trace and its observations
for trace in traces.data:
    trace_data = json.loads(trace.json())
    trace_id = trace.id
    
    # Fetch observations for this trace
    observations = langfuse.fetch_observations(trace_id=trace_id)
    
    # Add observations to trace data
    trace_data['observations'] = [json.loads(obs.json()) for obs in observations.data]
    
    # Append to our list
    traces_with_observations.append(trace_data)

# Export to JSON file
with open('langfuse_traces_with_observations.json', 'w') as f:
    json.dump(traces_with_observations, f, indent=2)

print(f"Exported {len(traces_with_observations)} traces with their observations to traces_with_observations.json")
```
